```python
## Function to create T0 - Hourly counter.

#number = 72 ## 12/24/72 hours
def count_T0(df): 
    df['T0'] = 0
    countFlag = 0
    Counter = 0
    count_no_sepsis = 0
    #Creating T0 going forward for sepsis and non sepsis patients
    for i, row in df.iterrows():
        if row['Temp Sepsis Flag'] == 1:
            countFlag = 1
        if countFlag==1:
            df.loc[i, 'T0'] = Counter
            Counter += 1
        if row['Sepsis Flag'] == 0 and df.loc[i, 'Counter'] >= 72:
            df.loc[i, 'T0'] = count_no_sepsis
            count_no_sepsis += 1
    
    df['Test'] = 0
    count = 0
    #Creating T0 going backward for sepsis and non sepsis patients
    for i, row in df.iterrows():
        #Counting the number of rows to know how to fill in from the first row and so on
        if df.loc[i, 'T0'] == 0:
            count += 1
        elif df.loc[i, 'T0'] == 1:
            #Counting
            for j, row in df.iterrows():
                df.loc[j, 'Test']= -count +1
                count -= 1
        else:
            df.loc[i, 'Test'] = df.loc[i, 'T0']
        df.loc[i, 'Test'] = df.loc[i, 'T0']
    
    df.drop("T0", axis = 1, inplace = True)
    df.rename({"Test": "T0"}, axis=1, inplace = True)
    return df

## Splitting each patient to a seperate dataframe
dfs = list()
for i in df_counter['h-num_demo'].unique():
    dfs.append(count_T0(df_counter[df_counter['h-num_demo']==i].copy()))    
    
## Combining all the dataframes of all the patients together to Full_DF
Full_DF = pd.concat(dfs)

## Dropping unnecessary columns which have been created during the manipulations
Full_DF = Full_DF.drop(["index"],axis=1)
Full_DF = Full_DF.drop(["Unnamed: 2"],axis=1)

display(Full_DF)
#Full_DF.to_csv('out.csv')
```
